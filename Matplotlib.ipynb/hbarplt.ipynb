```python
# HorizontalBarPlot.py
import matplotlib.pyplot as plt

# Data 
Animals = ['Cat', 'Bat', 'Ant', 'Deer']
values = [20, 40, 50, 25]

# horizontal bar plot
plt.figure(figsize=(8, 6))
plt.barh(Animals, values, color='green')

plt.xlabel('Values')
plt.ylabel('Animals')
plt.title('Horizontal Bar Plot Example')

plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.show()
```
