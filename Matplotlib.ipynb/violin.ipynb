```python
# ViolinPlot creation
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

#Generate synthetic data
np.random.seed(20)  
category_a = np.random.normal(70, 20, 120)  
category_b = np.random.normal(85, 15, 120)  
category_c = np.random.normal(60, 10, 120)   
data = [category_a, category_b, category_c]
labels = ['Category A', 'Category B', 'Category C']

#Create the violin plot
plt.figure(figsize=(10, 5))
sns.violinplot(data=data)
plt.xticks(ticks=range(len(labels)), labels=labels)

# Customize the plot
plt.title("Violin Plot Example", fontsize=20)
plt.ylabel("Values", fontsize=15)
plt.xlabel("Categories", fontsize=11)

plt.tight_layout()
plt.show()
```
