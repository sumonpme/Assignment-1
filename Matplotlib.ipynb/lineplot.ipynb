```python
# BasicLinePlot.py
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

# Data 
x = [1, 2, 3, 4, 5, 6]
y = [1, 4, 9, 16, 25, 36]

# Creating the line plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, marker='o', linestyle='-', color='green', label='y = x^2')
plt.title('Basic Line Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid()
plt.show()
```
