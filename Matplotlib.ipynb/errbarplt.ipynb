```python
# ErrorBarPlot.py
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 10)  
y = np.cos(x) + 0.5 * np.random.randn(10)  
y_err = 0.1  

# Creating the error bar plot
plt.figure(figsize=(8, 6))
plt.errorbar(x, y, yerr=y_err, fmt='o', ecolor='red', capsize=8, label='Data with Error')

plt.title('Error Bar Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

plt.show()
```
