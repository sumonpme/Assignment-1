```python
# Subplots.py
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 15, 100)
y1 = np.sin(x)

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))  

# subplot
axs[0, 0].plot(x, y1, color='green')
axs[0, 0].set_title("Sine Wave")
plt.tight_layout()

# Show the plots
plt.show()
```
