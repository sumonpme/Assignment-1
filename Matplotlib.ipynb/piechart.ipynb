```python
# PieChartExploded.py
import matplotlib.pyplot as plt

# Data for the pie chart
labels = ['Apple', 'Banana', 'Orange', 'Mango']
sizes = [25, 30, 35, 20]  
explode = (0, 0.1, 0, 0)  

# Creating the pie chart
plt.figure(figsize=(6, 6))
plt.pie(sizes, explode=explode, labels=labels, autopct='%1.0f%%', startangle=120)

plt.title('Exploded Pie Chart Example')
plt.show()
```
