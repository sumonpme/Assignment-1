```python
# DualAxisPlot.py
import matplotlib.pyplot as plt
import numpy as np

# Generating data
x = np.linspace(0, 20, 120)  
y1 = np.cos(x) 
y2 = np.exp(0.5 * x)  

fig, ax1 = plt.subplots(figsize=(8, 6))
ax2 = ax1.twinx()

ax1.plot(x, y1, 'g-')
ax2.plot(x, y2, 'b-')

ax1.set_xlabel('X-axis')
ax1.set_ylabel('Exponential', color='g')
ax2.set_ylabel('Logarithm', color='b')

plt.title('Dual Axis Plot')
plt.show()

```
