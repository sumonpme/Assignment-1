```python
# StepPlot.py
import matplotlib.pyplot as plt
import numpy as np

# data
x = np.arange(0, 10, 2)
y = np.log10(x + 1) 

# Create a step plot
plt.figure(figsize=(6, 4))
plt.step(x, y, where='mid', color='yellow', label='Log10 Values')

# Customize the plot
plt.title("Step Plot Example", fontsize=16)
plt.xlabel("Index", fontsize=10)
plt.ylabel("Value", fontsize=10)
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()

```
