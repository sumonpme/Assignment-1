```python
# StemPlot.py
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 10, 8)
y = np.tan(x)

# Create a stem plot
plt.figure(figsize=(8, 5))
plt.stem(x, y, linefmt='o-', markerfmt='m', basefmt='--')


plt.title("Stem Plot Example", fontsize=14)
plt.xlabel("X values", fontsize=12)
plt.ylabel("Y values", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)

# Show the plot
plt.tight_layout()
plt.show()
```
